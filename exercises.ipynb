```python
from random import randint, choice

def generate_random_product_dataset(num_products=100):
    categories = ["Electronics", "Home", "Sports", "Books", "Clothing", "Toys"]
    dataset = [
        {
            "name": f"Product {i}",
            "price": randint(10, 200), 
            "category": choice(categories), 
        }
        for i in range(1, num_products + 1)
    ]
    return dataset


def group_by_category(products):
    grouped = {}
    for product in products:
        category = product["category"]
        price = product["price"]
        if category not in grouped:
            grouped[category] = []
        grouped[category].append(price)
    return grouped

def calculate_high_avg_categories(products):
    grouped = group_by_category(products)
    
    category_avg = list(
        map(
            lambda category: {
                "category": category,
                "average_price": sum(grouped[category]) / len(grouped[category])
            },
            grouped
        )
    )
    
    high_avg_categories = list(
        filter(
            lambda item: item["average_price"] > 50,
            category_avg
        )
    )
    
    return high_avg_categories


random_product_dataset = generate_random_product_dataset(50)
result = calculate_high_avg_categories(random_product_dataset)
print(result)
```

```python
def generate_random_employee_dataset(num_employees=50):
    departments = ["Engineering", "HR", "Marketing", "Sales", "Finance", "IT"]
    dataset = [
        {
            "name": f"Employee {i}",
            "salary": randint(40000, 120000),  
            "department": choice(departments) 
        }
        for i in range(1, num_employees + 1)
    ]
    return dataset


def group_salaries_by_department(employees):
    grouped = {}
    for employee in employees:
        department = employee["department"]
        salary = employee["salary"]
        if department not in grouped:
            grouped[department] = []
        grouped[department].append(salary)
    return grouped

def calculate_high_avg_departments(employees, threshold=65000):
    grouped = group_salaries_by_department(employees)
    department_avg = list(
        map(
            lambda department: {
                "department": department,
                "average_salary": sum(grouped[department]) / len(grouped[department])
            },
            grouped
        )
    )
    high_avg_departments = list(
        filter(
            lambda item: item["average_salary"] > threshold,
            department_avg
        )
    )

    return high_avg_departments

random_employee_dataset = generate_random_employee_dataset(50)
result_random = calculate_high_avg_departments(random_employee_dataset, threshold=65000)
print("Random dataset result:")
print(result_random)
```

# Biopython

Write the following five functions to analyze global alignments between two sequences using Biopython's `pairwise2` module:

1. **countMatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment (pairwise2.globalxx) of the same length. It returns the number of positions where the elements of both sequences match.

2. **countMismatches(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of positions where the elements of the two sequences are different (i.e., they are not gaps, and the characters do not match).

3. **countGapOpens(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap openings in the alignment (a gap is opened when a '-' appears in the sequence).

4. **countGapExtensions(s1, s2)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment of the same length. It returns the number of gap extensions (where '-' continues in the alignment after an initial gap is opened).

5. **getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty)**  
   This function takes two sequences (`s1`, `s2`) aligned using global alignment and returns the alignment score based on the provided scoring scheme: `matchScore` for matches, `mismatchPenalty` for mismatches, `gapOpenPenalty` for opening a gap, and `gapExtensionPenalty` for extending a gap.

```python
from Bio import pairwise2
def countMatches(s1, s2):
    count = 0
    for a, b in zip(s1, s2):
        if a == b and a != '-' and b != '-': 
            count += 1
    return count

def countMismatches(s1, s2):
    count = 0
    for a, b in zip(s1, s2):
        if a != b and a != '-' and b != '-':
            count += 1
    return count

def countGapOpens(s1, s2):
    count = 0
    gap_opened = False
    for a, b in zip(s1, s2):
        if a == '-' or b == '-':
            if not gap_opened:
                count += 1
                gap_opened = True
        else:
            gap_opened = False
    return count

def countGapExtensions(s1, s2):
    count = 0
    gap_opened = False
    for a, b in zip(s1, s2):
        if a == '-' or b == '-':
            if gap_opened:
                count += 1
            else:
                gap_opened = True
        else:
            gap_opened = False
    return count

def getScore(s1, s2, matchScore, mismatchPenalty, gapOpenPenalty, gapExtensionPenalty):
    alignments = pairwise2.align.globalxx(s1, s2)
    alignment = alignments[0]
    aligned_s1, aligned_s2 = alignment[:2]
    score = 0
    for a, b in zip(aligned_s1, aligned_s2):
        if a == b and a != '-':  
            score += matchScore
        elif a != b and a != '-' and b != '-':  
            score += mismatchPenalty
        elif a == '-' or b == '-': 
            if aligned_s1.find('-') == 0 or aligned_s2.find('-') == 0: 
                score += gapOpenPenalty
            else: 
                score += gapExtensionPenalty
    
    return score


s1 = "AGCTG"
s2 = "AGCT-"
match_score = 2
mismatch_penalty = -1
gap_open_penalty = -2
gap_extension_penalty = -1


alignment = pairwise2.align.globalxx(s1, s2)[0]


print("Matches:", countMatches(alignment[0], alignment[1]))
print("Mismatches:", countMismatches(alignment[0], alignment[1]))
print("Gap Opens:", countGapOpens(alignment[0], alignment[1]))
print("Gap Extensions:", countGapExtensions(alignment[0], alignment[1]))


score = getScore(s1, s2, match_score, mismatch_penalty, gap_open_penalty, gap_extension_penalty)
print("Alignment score:", score)
```
